# Indexing with boolean arrays

```python
import random
```

```python
# it works like masking
arr4=np.random.randint(low=20,high=80,size=24).reshape(4,6)
```

```python
arr4
```

```python
arr4[0][2]
```

```python
arr4>60
```

```python
arr4[arr4>60]
```

```python
arr4[(arr4>60)&(arr4%2!=0)]
```

## Numpy Operations

```python
arr1=np.array([1,2,3,4,5,6])
arr2=np.array([5,6,7,8,9,10])
```

```python
arr1-arr2
```

```python
arr1+arr2
```

```python
arr1*arr2
```

```python
# scaler multiplication
arr1*2
```

```python
arr1>4
```
